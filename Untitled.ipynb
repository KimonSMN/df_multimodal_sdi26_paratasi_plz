```python
import pandas as pd
import numpy as np
import tarfile
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# ΒΗΜΑ 1: Εύρεση των Top-5 Genres & Αρχικό Φιλτράρισμα
# ---------------------------------------------------------
df_genres = pd.read_csv('id_genres.csv', sep='\t')

# Βρίσκουμε τα 5 πιο συχνά είδη
top5_genres = df_genres['genres'].value_counts().head(5).index.tolist()
print(f"Τα Top-5 Genres είναι: {top5_genres}")

# Κρατάμε μόνο τις γραμμές που ανήκουν στα Top-5
df_genres_filtered = df_genres[df_genres['genres'].isin(top5_genres)]

# Δημιουργούμε ένα Set με τα valid IDs για πολύ γρήγορη αναζήτηση (O(1) complexity)
target_ids = set(df_genres_filtered['id'])


# ---------------------------------------------------------
# ΒΗΜΑ 2: Φόρτωση Audio Features με Chunksize (Memory Optimization)
# ---------------------------------------------------------
mfcc_chunks = []

# Διαβάζουμε το τεράστιο αρχείο τμηματικά (ανά 500 γραμμές)
for chunk in pd.read_csv('id_mfcc_stats.tsv.bz2', sep='\t', compression='bz2', chunksize=500):
    # Κρατάμε από το chunk ΜΟΝΟ τα IDs που ανήκουν στα Top-5 Genres
    filtered_chunk = chunk[chunk['id'].isin(target_ids)]
    mfcc_chunks.append(filtered_chunk)

# Ενώνουμε τα καθαρισμένα chunks σε ένα DataFrame
df_mfcc_filtered = pd.concat(mfcc_chunks, ignore_index=True)

# Ανανεώνουμε τα target_ids: Κρατάμε την τομή (Intersection), δηλαδή όσα έχουν ΚΑΙ Genre ΚΑΙ Audio
target_ids = target_ids.intersection(set(df_mfcc_filtered['id']))


# ---------------------------------------------------------
# ΒΗΜΑ 3: Στοχευμένη Εξαγωγή Στίχων (Targeted Extraction)
# ---------------------------------------------------------
lyrics_data = []

with tarfile.open('processed_lyrics.tar.gz', 'r:gz') as tar:
    for member in tar.getmembers():
        if member.isfile() and member.name.endswith('.txt'):
            song_id = member.name.split('/')[-1].replace('.txt', '')
            
            # ΕΞΟΙΚΟΝΟΜΗΣΗ ΧΡΟΝΟΥ/ΜΝΗΜΗΣ: Ανοίγουμε το αρχείο ΜΟΝΟ αν το χρειαζόμαστε
            if song_id in target_ids:
                f = tar.extractfile(member)
                if f:
                    text = f.read().decode('utf-8', errors='ignore').strip()
                    lyrics_data.append({'id': song_id, 'lyrics': text})

df_lyrics = pd.DataFrame(lyrics_data)

# Ανανεώνουμε τα target_ids για τελευταία φορά: Πρέπει να έχουν Genre, Audio ΚΑΙ Στίχους (Intersection και των 3)
valid_ids = target_ids.intersection(set(df_lyrics['id']))


# ---------------------------------------------------------
# ΒΗΜΑ 4: Φόρτωση Μεταδεδομένων & Τελικό Inner Merge
# ---------------------------------------------------------
df_info = pd.read_csv('id_information.csv', sep='\t')
df_tags = pd.read_csv('id_tags.csv', sep='\t')

# Φιλτράρουμε τα πάντα με βάση τα valid_ids ΠΡΙΝ το merge
df_info = df_info[df_info['id'].isin(valid_ids)]
df_genres_filtered = df_genres_filtered[df_genres_filtered['id'].isin(valid_ids)]
df_tags = df_tags[df_tags['id'].isin(valid_ids)]
df_mfcc_filtered = df_mfcc_filtered[df_mfcc_filtered['id'].isin(valid_ids)]
df_lyrics = df_lyrics[df_lyrics['id'].isin(valid_ids)]

# Κάνουμε Inner Merge (Τομή) σε όλα, χρησιμοποιώντας το 'id' ως κλειδί
master_df = df_info.merge(df_genres_filtered, on='id', how='inner')
master_df = master_df.merge(df_tags, on='id', how='left') # Χρησιμοποιούμε left για τα tags, καθώς ίσως λείπουν κάποια
master_df = master_df.merge(df_lyrics, on='id', how='inner')
master_df = master_df.merge(df_mfcc_filtered, on='id', how='inner')

# Διαχειριζόμαστε τα πιθανά κενά στα tags
if 'tags' in master_df.columns:
    master_df['tags'] = master_df['tags'].fillna('')

# ---------------------------------------------------------
# ΤΕΛΙΚΟΣ ΕΛΕΓΧΟΣ
# ---------------------------------------------------------
print("Missing values (πρέπει να είναι μηδενικά για lyrics, genres, mfcc):\n", master_df.isna().sum().head(10))
print(f"\nΤελικές διαστάσεις καθαρού Dataset: {master_df.shape}")
master_df.head(20)
```
