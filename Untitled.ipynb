```python
# Συλλογή Δεδομένων - Filtering

import pandas as pd

# Δημιουργουμε ενα DataFrame από το αρχείο id_genres.csv με seperator το tab
genres_dataFrame = pd.read_csv('id_genres.csv', sep='\t') 

# Κανουμε το dataframe λιστα με τα genres για να μπορουμε να τα επεξεργαστουμε
genres_dataFrame['genres'] = genres_dataFrame['genres'].str.split(',')

# Κανουμε explode για να εχουμε μια γραμμη για καθε genre
exploded_genres = genres_dataFrame.explode('genres')

# top 5 genres με βαση το ποσα τραγουδια εχουν καθε genre
# value_counts() μετραει ποσα τραγουδια εχει καθε genre και nlargest(5) κραταει τα 5 πιο συχνα
top_5_series = exploded_genres['genres'].value_counts().nlargest(5)
# Μετατρεπουμε το Series σε λιστα
top_5_genres = top_5_series.index.tolist()

print(f"Top 5 Genres: {top_5_genres}")

# Κραταμε στο filtered_genres μονο τα τραγουδια που ειναι στα το genre στα top 5
filtered_genres = exploded_genres[exploded_genres['genres'].isin(top_5_genres)]

# Αφαιρουμε τα 2πλα genres για καθε τραγουδι, κρατωντας το πρωτο που εμφανιζεται
filtered_genres = filtered_genres.drop_duplicates(subset=['id'], keep='first')

# Ονομαζουμε το genres -> genre διοτι τωρα εχουμε μονο ενα genre
filtered_genres = filtered_genres.rename(columns={'genres': 'genre'})[['id', 'genre']]

# Set για πιο γρηγορη αναζητηση των id των τραγουδιων που θα επεξεργαστουμε
target_ids = set(filtered_genres['id'])
print(f"Θα επεξεργαστουμε {len(target_ids)} τραγουδια.")
```

```python
# Συλλογή Δεδομένων - Intersection & Αποτελεσμα

import tarfile

audio_chunks = []

# Διαβάζουμε το αρχείο id_mfcc_stats.tsv.bz2 σε chunks (δες εκφώνηση)
mfcc_reader = pd.read_csv('id_mfcc_stats.tsv.bz2', sep='\t', chunksize=1000)

# Για καθε chunk, φιλτραρουμε μονο τα τραγουδια που ειναι στα target_ids και τα προσθετουμε στη λιστα audio_chunks
for chunk in mfcc_reader:
    filtered_chunk = chunk[chunk['id'].isin(target_ids)]
    audio_chunks.append(filtered_chunk)
# Ενωνουμε ολα τα φιλτραρισμενα chunks σε ενα DataFrame
audio_df = pd.concat(audio_chunks, ignore_index=True)

# Διαβαζουμε το processed_lyrics.tar.gz και για καθε αρχειο που ειναι στα target_ids, διαβαζουμε τα lyrics και τα προσθετουμε σε μια λιστα
lyrics_list = []
with tarfile.open('processed_lyrics.tar.gz', 'r:gz') as tar:
    for member in tar.getmembers():
        if member.isfile() and member.name.endswith('.txt'):

            # Το song_id ειναι το ονομα του αρχειου χωρις .txt
            song_id = member.name.split('/')[-1].replace('.txt', '')
            
            # Αν το song_id ειναι στα target_ids, διαβαζουμε τα lyrics και τα προσθετουμε στη λιστα
            if song_id in target_ids:
                f = tar.extractfile(member)
                if f:
                    content = f.read().decode('utf-8')
                    lyrics_list.append({'id': song_id, 'lyrics': content})

# Δημιουργουμε ενα DataFrame από τη λιστα με τα lyrics
lyrics_df = pd.DataFrame(lyrics_list)

# Κανουμε merge τα 3 DataFrames με βαση το id
final_df = filtered_genres.merge(lyrics_df, on='id').merge(audio_df, on='id')

# Βγαζουμε το id απο τις στηλες mfcc
mfcc_cols = [c for c in audio_df.columns if c != 'id']
# DataFrame με νεα στηλη mfcc_stats που καθε γραμμη ειναι μια λιστα με τα stats
final_df['mfcc_stats'] = final_df[mfcc_cols].values.tolist()

# Κραταμε μονο τις στηλες που θελουμε για το τελικο DataFrame
final_df = final_df[['id', 'genre', 'lyrics', 'mfcc_stats']]

#final_df.to_csv('final_music_data.csv', index=False) # για csv 

print(f"Dataset intersection complete with: {len(final_df)} songs.")
```

```python
import pandas as pd
import numpy as np
import tarfile
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
import nltk

# Προετοιμασία NLTK
nltk.download('vader_lexicon')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Διαχειρηση δεδομενων (απο προηγουμενα βηματα)

# Φορτωση και φιλτραρισμα των δεδομενων
genres_df = pd.read_csv('id_genres.csv', sep='\t') 
genres_df['genres_list'] = genres_df['genres'].str.split(',')
exploded_genres = genres_df.explode('genres_list')
top_5_genres = exploded_genres['genres_list'].value_counts().nlargest(5).index.tolist()
filtered_genres = exploded_genres[exploded_genres['genres_list'].isin(top_5_genres)].drop_duplicates(subset=['id'])
target_ids = set(filtered_genres['id'])

audio_list = []
mfcc_reader = pd.read_csv('id_mfcc_stats.tsv.bz2', sep='\t', chunksize=1000)
for chunk in mfcc_reader:
    audio_list.append(chunk[chunk['id'].isin(target_ids)])
audio_df = pd.concat(audio_list, ignore_index=True)

lyrics_data = []
with tarfile.open('processed_lyrics.tar.gz', 'r:gz') as tar:
    for member in tar.getmembers():
        if member.isfile() and member.name.endswith('.txt'):
            song_id = member.name.split('/')[-1].replace('.txt', '')
            if song_id in target_ids:
                f = tar.extractfile(member)
                if f: lyrics_data.append({'id': song_id, 'lyrics': f.read().decode('utf-8')})
lyrics_df = pd.DataFrame(lyrics_data)

final_df = filtered_genres.rename(columns={'genres_list': 'genre'}).merge(lyrics_df, on='id').merge(audio_df, on='id')

# Προετιμασια για dimensionality reduction 

# Word2Vec για text embeddings
sentences = [lyric.lower().split() for lyric in final_df['lyrics']]
w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)
def get_avg_w2v(tokens, model, size):
    vecs = [model.wv[word] for word in tokens if word in model.wv]
    return np.mean(vecs, axis=0) if vecs else np.zeros(size)
text_embeddings = np.array([get_avg_w2v(s, w2v_model, 100) for s in sentences])

# PCA για audio embeddings
mfcc_cols = [c for c in audio_df.columns if c != 'id']
audio_scaled = StandardScaler().fit_transform(final_df[mfcc_cols])
audio_embeddings = PCA(n_components=0.95).fit_transform(audio_scaled)

# world cloud & bar chart

tags_df = pd.read_csv('id_tags.csv', sep='\t')
final_tags = final_df[['id', 'genre']].merge(tags_df, on='id')

def plot_tags_eda(genre_name):
    genre_data = final_tags[final_tags['genre'] == genre_name]
    all_tags = ','.join(genre_data['tags'].dropna()).split(',')
    
    # WordCloud
    wc = WordCloud(width=800, height=400, background_color='white').generate(' '.join(all_tags))
    plt.figure(figsize=(10, 4)); plt.imshow(wc); plt.title(f"WordCloud: {genre_name}"); plt.axis('off'); plt.show()
    
    # Bar Chart
    top_10_tags = pd.Series(all_tags).value_counts().nlargest(10)
    plt.figure(figsize=(10, 4)); sns.barplot(x=top_10_tags.values, y=top_10_tags.index)
    plt.title(f"Top 10 Tags: {genre_name}"); plt.show()

# εκτελεση για το πιο συχνο και το πιο σπανιο απο τα top 5 genres
plot_tags_eda(top_5_genres[0])
plot_tags_eda(top_5_genres[-1])

# Ποσα τραγουδια ανοικουν σε 1, 2, 3... genres
genre_counts = genres_df['genres_list'].apply(len).value_counts().sort_index()
plt.figure(figsize=(8, 4)); sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title("Πληθος genre στα οποια ανηκει καθε τραγουδι"); plt.show()

# Αντιστροφο
top_n_dist = exploded_genres['genres_list'].value_counts().nlargest(10)
plt.figure(figsize=(10, 4)); sns.barplot(x=top_n_dist.index, y=top_n_dist.values)
plt.title("Κατανομη τραγουδιων στα top 10 genres"); plt.xticks(rotation=45); plt.show()

# Μεγεθος στιχων και stop words
final_df['word_count'] = final_df['lyrics'].apply(lambda x: len(x.split()))
final_df['meaningful_words'] = final_df['lyrics'].apply(lambda x: len([w for w in x.lower().split() if w not in stop_words]))

plt.figure(figsize=(10, 5))
sns.histplot(final_df['word_count'], color="skyblue", label='Συνολικές Λέξεις', kde=True)
sns.histplot(final_df['meaningful_words'], color="red", label='Meaningful Λέξεις', kde=True)
plt.legend(); plt.title("Κατανομη μεγεθους στοιχων"); plt.show()

#VADER Sentiment
sia = SentimentIntensityAnalyzer()
final_df['sentiment'] = final_df['lyrics'].apply(lambda x: sia.polarity_scores(x)['compound'])
plt.figure(figsize=(10, 6)); sns.boxplot(x='genre', y='sentiment', data=final_df)
plt.title("VADER"); plt.show()

# 2D Scatter Plots
pca_2d = PCA(n_components=2)
text_2d = pca_2d.fit_transform(text_embeddings)
audio_2d = pca_2d.fit_transform(audio_embeddings)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
sns.scatterplot(x=text_2d[:,0], y=text_2d[:,1], hue=final_df['genre'], ax=ax1, s=15, alpha=0.7)
ax1.set_title("Text Embeddings 2D (Clusters)"); 
sns.scatterplot(x=audio_2d[:,0], y=audio_2d[:,1], hue=final_df['genre'], ax=ax2, s=15, alpha=0.7)
ax2.set_title("Audio Embeddings 2D (Clusters)"); plt.show()

#Cosine Similarity
def get_similar_songs(idx, embeddings, df, mode_name):
    dist = cosine_similarity([embeddings[idx]], embeddings)[0]
    sim_indices = dist.argsort()[-6:-1][::-1] # Τα 5 πιο όμοια (εκτός του εαυτού του)
    print(f"\nTop 5 similar songs to '{df.iloc[idx]['id']}' based on {mode_name}:")
    print(df.iloc[sim_indices][['id', 'genre']])

# πχ για το πρωτο τραγουδι
get_similar_songs(0, text_embeddings, final_df, "Lyrics")
get_similar_songs(0, audio_embeddings, final_df, "Audio")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity

# --- ΛΥΣΗ ΓΙΑ ΤΟ PERFORMANCE WARNING ---
final_df = final_df.copy()

# ==========================================
# 1. Bar Chart: Τα 10 πιο συχνά tags ανά είδος 
# ==========================================
print("--- Top 10 Tags ανά Genre ---")
for genre in top_5_genres:
    # Βρίσκουμε τα tags για το συγκεκριμένο είδος
    genre_data = final_tags[final_tags['genre'] == genre]
    
    # Ενώνουμε όλα τα tags και τα χωρίζουμε με κόμμα
    all_tags = ','.join(genre_data['tags'].dropna()).split(',')
    # Καθαρίζουμε από κενά spaces
    all_tags = [tag.strip() for tag in all_tags if tag.strip()]
    
    # Μετράμε και κρατάμε τα top 10
    top_10 = pd.Series(all_tags).value_counts().head(10)
    
    plt.figure(figsize=(8, 4))
    sns.barplot(x=top_10.values, y=top_10.index, palette='viridis')
    plt.title(f"Top 10 Tags for {genre}")
    plt.xlabel("Frequency")
    plt.ylabel("Tags")
    plt.tight_layout()
    plt.show()

# ==========================================
# 2. Πόσα τραγούδια κατατάσσονται στο καθένα από τα Top-5 είδη 
# ==========================================
plt.figure(figsize=(8, 5))
top_genres_counts = final_df['genre'].value_counts()
sns.barplot(x=top_genres_counts.index, y=top_genres_counts.values, palette='magma')
plt.title("Αριθμός Τραγουδιών ανά Top-5 Είδος")
plt.xlabel("Είδος (Genre)")
plt.ylabel("Αριθμός Τραγουδιών")
plt.show()

# ==========================================
# 3. Κατανομή Μεγέθους Στίχων 
# ==========================================
# Πλήθος χαρακτήρων
final_df['char_count'] = final_df['lyrics'].apply(len)
# Πλήθος λέξεων (split με κενά)
final_df['word_count'] = final_df['lyrics'].apply(lambda x: len(str(x).split()))
# Πλήθος meaningful λέξεων (χωρίς stopwords)
final_df['meaningful_word_count'] = final_df['lyrics'].apply(
    lambda x: len([w for w in str(x).lower().split() if w not in stop_words])
)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(final_df['char_count'], bins=50, ax=axes[0], color='skyblue')
axes[0].set_title('Κατανομή Χαρακτήρων')

sns.histplot(final_df['word_count'], bins=50, ax=axes[1], color='lightgreen')
axes[1].set_title('Κατανομή Λέξεων')

sns.histplot(final_df['meaningful_word_count'], bins=50, ax=axes[2], color='salmon')
axes[2].set_title('Κατανομή Meaningful Λέξεων (No Stopwords)')

plt.tight_layout()
plt.show()

# ==========================================
# 4. Similarity Analysis (Συνημιτονοειδής Ομοιότητα) [cite: 37]
# ==========================================
# Επιλέγουμε τυχαία ένα τραγούδι ως "στόχο" (π.χ. το πρώτο, index = 0)
target_idx = 0
target_id = final_df.iloc[target_idx]['id']
target_genre = final_df.iloc[target_idx]['genre']
print(f"\nΑναζήτηση ομοιοτήτων για το Τραγούδι-Στόχο: {target_id} (Genre: {target_genre})\n")

# --- Similarity βάσει Στίχων (Text) ---
text_sims = cosine_similarity(text_embeddings[target_idx].reshape(1, -1), text_embeddings)[0]
# argsort() επιστρέφει τους δείκτες σε αύξουσα σειρά. Θέλουμε τα 5 μεγαλύτερα εκτός από το ίδιο το τραγούδι
top5_text_indices = text_sims.argsort()[-6:-1][::-1]

print(">>> Top 5 Όμοια Τραγούδια βάσει TEXT (Στίχοι):")
for idx in top5_text_indices:
    print(f"ID: {final_df.iloc[idx]['id']} | Genre: {final_df.iloc[idx]['genre']} | Similarity: {text_sims[idx]:.4f}")

# --- Similarity βάσει Ήχου (Audio) ---
audio_sims = cosine_similarity(audio_embeddings[target_idx].reshape(1, -1), audio_embeddings)[0]
top5_audio_indices = audio_sims.argsort()[-6:-1][::-1]

print("\n>>> Top 5 Όμοια Τραγούδια βάσει AUDIO (MFCCs):")
for idx in top5_audio_indices:
    print(f"ID: {final_df.iloc[idx]['id']} | Genre: {final_df.iloc[idx]['genre']} | Similarity: {audio_sims[idx]:.4f}")
```
